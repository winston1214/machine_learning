```
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
def lag_finder(y1, y2):
    n = len(y1)
    corr = signal.correlate(y2, y1, mode='same',method='fft') / np.sqrt(signal.correlate(y1, y1, mode='same',method='fft')[int(n/2)] \
                                                           * signal.correlate(y2, y2, mode='same',method='fft')[int(n/2)])
    return corr
```

```
corr = lag_finder(x['X (G)'],x['Velocity (Km/h)'])
plt.plot(corr)
plt.ylabel('corr')
plt.xlabel('Time')
plt.title('X - Velocity')
```

```
corr = lag_finder(x['Y (G)'],x['Velocity (Km/h)'])
plt.plot(corr)
plt.ylabel('corr')
plt.xlabel('Time')
plt.title('Y - Velocity')
```

```
corr = lag_finder(x['Z (G)'],x['Velocity (Km/h)'])
plt.plot(corr)
plt.ylabel('corr')
plt.xlabel('Time')
plt.title('Z - Velocity')
```

```
plt.figure(figsize=(20,12))
mask = np.zeros_like(x[['X (G)','Y (G)','Z (G)','Velocity (Km/h)']].corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(x[['X (G)','Y (G)','Z (G)','Velocity (Km/h)']].corr(), cmap=sns.diverging_palette(20, 220, n=200), annot=True, center = 0, mask=mask); 
plt.title("X-Y-Z-Velocity", fontsize = 30)
```
