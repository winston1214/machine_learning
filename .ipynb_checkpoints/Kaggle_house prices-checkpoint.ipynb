```python
# missing values
ms = pd.DataFrame(columns=['col','missing'])
idx=0
for i in range(data.shape[1]):
    if data.isnull().sum()[i]>0:
        ms.loc[idx,'col'] = data.isnull().sum().index[i]
        ms.loc[idx,'missing'] = data.isnull().sum()[i]/data.shape[0] * 100
        idx+=1
    else:
        continue
ms = ms.sort_values(by= 'missing',ascending=False)
ms
```

# Impute missing values

```python
col = ['PoolQC','MiscFeature','Alley','Fence','FireplaceQu','GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
      'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2','MasVnrType','MSSubClass']
for i in col:
    data[i] = data[i].fillna('None')
```

```python
data["LotFrontage"] = data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

```python
col2 = ['GarageYrBlt', 'GarageArea', 'GarageCars','BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath',
        'BsmtHalfBath','MasVnrArea']
for i in col2:
    data[i] = data[i].fillna(0)
```

```python
col3 = ['MSZoning','Electrical','KitchenQual','Exterior1st','Exterior2nd','SaleType']
for i in col3:
    data[i] = data[i].fillna(data[i].mode()[0])
```

```python
data['Functional'] = data['Functional'].fillna('Typ')
```

```python
data.drop('Utilities',axis=1,inplace=True)
```

```python
ms = pd.DataFrame(columns=['col','missing'])
idx=0
for i in range(data.shape[1]):
    if data.isnull().sum()[i]>0:
        ms.loc[idx,'col'] = data.isnull().sum().index[i]
        ms.loc[idx,'missing'] = data.isnull().sum()[i]/data.shape[0] * 100
        idx+=1
    else:
        continue
ms = ms.sort_values(by= 'missing',ascending=False)
ms
```

we can't find missing values

```python
col = ['MSSubClass','OverallCond','YrSold','MoSold']
for i in col:
    data[i] = data[i].astype(str)
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ['FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 
        'YrSold', 'MoSold']
# LabelEncoder to categorical features
le = LabelEncoder()
for i in cols:
    data[i] = le.fit_transform(data[i].values)
print(data.shape)
```

```python
data['TotalSF'] = data['TotalBsmtSF'] + data['1stFlrSF'] + data['2ndFlrSF']
```

### box-cox 변환
- 정규화가 아닌 자료를 정규화 시키기 위한 방법
- boxcox1p 를 사용하는 이유는 saleprice도 log1p를 사용했기 때문에 사용

```python
numeric_feats = data.dtypes[data.dtypes != "object"].index

# Check the skew of all numerical features
skewed_feats = data[numeric_feats].apply(lambda x: skew(x.dropna()))
skewed = pd.DataFrame({'skew':skewed_feats}).sort_values(by='skew',ascending=False)
skewed.head()
```

```python
from scipy.special import boxcox1p
skewed = skewed[abs(skewed)>0.75]
lam = 0.15
for col in skewed.index:
    data[col] = boxcox1p(data[col],lam)
```
